```python
from databricks.sdk import WorkspaceClient
import boto3
```

```python
w = WorkspaceClient()

tempCredential = w.credentials.generate_temporary_service_credential("one-env-databricks-access")

session = boto3.Session(
    aws_access_key_id=tempCredential.aws_temp_credentials.access_key_id,
    aws_secret_access_key=tempCredential.aws_temp_credentials.secret_access_key,
    aws_session_token=tempCredential.aws_temp_credentials.session_token,
    region_name='us-east-2'
)

s3_client=session.client('s3')
```
